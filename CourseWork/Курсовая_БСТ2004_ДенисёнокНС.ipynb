### Задание 131А
### cAPS lOCK:
зАЧЕМ НУЖНА КЛАВИША cAPS lOCK?

Caps Lock — клавиша компьютерной клавиатуры, предназначенная для автоматической (постоянной) смены регистра букв со строчных на прописные. Будучи случайно нажатой, она приводит к последствиям вроде первого абзаца в условии этой задачи.

Будем считать, что слово набрано с ошибочно нажатой клавишей Caps Lock, если:

- либо оно полностью состоит из прописных букв;
- либо прописными являются все его буквы, кроме первой.
В таком случае, нужно автоматически поменять регистр всех букв на противоположный. Например, регистр букв слов «hELLO», «HTTP», «z» должен быть изменен.

Напишите программу, которая применяет описанное выше правило или оставляет слово без изменения, если оно не применимо.

**Входные данные** <br />
В первой строке входных данных записано слово, состоящее из прописных или строчных букв латинского алфавита. Длина слова — от 1 до 100 символов включительно.

**Выходные данные** <br />
Выведите результат обработки данного слова.

### Выполнение:

```python
#text = input() #ввод текста
def cAPS(text):
    if len(text) > 0:
        swapcase = True
        for i in range (1, len(text)): #пробегаемся по символам, начиная со второго
            if text[i].islower(): #если далее идут строчные символы, то менять ничего не нужно
                swapcase = False
        if swapcase:
            text = text[0] +text[1:].lower() #приклеиваем к первому символу остальную строку в строчном виде
            if text[0].isupper(): #если первый символ уже был прописным (и все остальные), то делаем его строчным
                text = text[0].lower() + text[1:]
            elif text[0].islower(): # если первый символ строчный (симптом зажатого caps), делаем его прописным
                text = text[0].upper() + text[1:]
    print(text)
cAPS("cAPS") # Expected Caps
cAPS("Lock") # Expected Lock
cAPS("CapsLock") # Expected CapsLock
```

«Игра 23» — вот во что сейчас играет Поликарп. Изначально у него есть целое число n и его цель преобразовать его в целое число m. За один ход он может умножить n на 2 или умножить n на 3. Он может совершать любое количество ходов.

Выведите количество ходов, сколько необходимо Поликарпу, чтобы получить из числа n число m. Выведите -1, если это невозможно сделать.

Легко доказать, что любой способ преобразования из числа n в число m содержит одинаковое количество ходов (т.е. количество ходов не зависит от способа преобразования).

**Входные данные**<br />
В единственной строке входных данных записаны два целых числа n и m (1≤n≤m≤5⋅108).

**Выходные данные**<br />
Выведите количество ходов, чтобы получить из числа n число m. Выведите -1, если это невозможно сделать. 

### Выполнение:

```python
#a, b = input().split(" ")

#if a != "" and b != "" and a.isdigit() and b.isdigit():
    #initial = int(a)
    #res = int(b)

def game23(initial, res):
    if initial != 0 and res != 0 and ((initial > 0 and res > 0) or(initial < 0 and res < 0)): #проверка на правильность введенного числа
        if res % 2 != 0 and res % 3 != 0 and res != initial: #Если число не делится на 2 и на 3, сразу можно выводить -1
            print(-1)
        else:
            print(steps(initial, res)) #Если делится на 2 или на 3, можно начинать считать количество шагов
    else:
        print(-1)

def steps(init, res):
    count = 0
    if init == res: #случай, когда ввели два одинаковых числа
        return 0
    while init < res:
        if res % (init * 2) == 0: #если остаток от деления желаемого числа на удвоенное текущее равен нулю, то умножаем на два
            init *= 2
        else: #если остаток желаемого деленного на утроенное текущее равен нулю, то умножаем на три
            init *= 3
        count += 1
    if init == res:
        return count
    return -1

game23(120, 51840) # Expected 7
game23(42, 42) # Expected 0
game23(48, 72) # Expected -1
game23(1, 6) # Expected 2
```

### Задание 1237А 
### Сбалансированные изменения рейтинга:
Только что завершился очередной раунд на Codeforces! Раунд собрал n участников. В соответствии с результатами, ожидаемое изменение рейтинга участника i равно ai. Эти изменения рейтинга идеально сбалансированы — их сумма равна 0.

К сожалению, из-за небольших технических проблем раунд был объявлен полурейтинговым. Это значит, что все изменения рейтинга должны быть поделены на два.

Однако есть два условия:

- Для каждого участника i обновлённое изменение рейтинга bi должно быть целым числом, максимально близким к ai/2. Это значит, что либо bi=⌊ai/2⌋, либо bi=⌈ai/2⌉. В частности, если ai чётно, то bi=ai/2. Здесь ⌊x⌋ обозначает округление вниз до наибольшего целого числа, не большего x, а ⌈x⌉ — округление вверх до наименьшего целого числа, не меньшего x.
- Обновлённые изменения рейтинга должны быть идеально сбалансированы — их сумма должна равняться 0.
Можете ли вы помочь с этой задачей?

**Входные данные**<br />
Первая строка содержит целое число n (2≤n≤13845) — число участников раунда.

Каждая из следующих n строк содержит целое число ai (−336≤ai≤1164) — ожидаемое изменение рейтинга i-го участника.

Сумма всех ai равна 0.

**Выходные данные**<br />
Выведите n целых чисел bi, каждое из которых равно обновлённому изменению рейтинга i-го участника в порядке ввода.

Для всех i должно быть верно либо bi=⌊ai/2⌋, либо bi=⌈ai/2⌉. Сумма всех bi должна быть равна 0.

Если есть несколько решений, выведите любое. Можно показать, что решение существует для любого корректного ввода. 

### Выполнение:

```python
import math
#n = int(input())
grades = []
def balanced_rating(grades):
    flag = True
    for i in grades:
        if i % 2 != 0: #если оценка была нечетной
            if flag == True: #и флаг равен истине, то делим на два и округляем вверх и устанавливаем флаг в ложь
                i = math.ceil(i/2)
                flag = False
            else: # если флаг ложь, то делим на два и округляем вниз и устанавливаем флаг в истину
                i = math.floor(i/2)
                flag = True
        else:
            i = int(i / 2) #если оценка была четной, то просто делим на два
        print(i)
balanced_rating([10, -5, -5]) #5 -2 -3
print("===========")
balanced_rating([-7, -29, 0, 3, 24, -29, 38]) #-3 -15 0 2 12 -15 19
```

В связи с началом учебного года во многих общежитиях возникли проблемы с поселением студентов. В одном из таких общежитий есть чудо-комната размером a × b квадратных метров. Но, по закону, на одного жильца должно быть не меньше 6 квадратных метров жилой площади (то есть, комната для n студентов должна иметь площадь не менее 6n квадратных метров). Комендант может увеличить любую из сторон комнаты (возможно, обе) на произвольное целое положительное количество метров. Помогите коменданту изменить комнату так, чтобы все n студентов могли жить в комнате, и итоговая площадь комнаты была как можно меньше.

**Входные данные**<br />
В единственной строке через пробел записано три целых числа n, a и b (1 ≤ n, a, b ≤ 10^9) — количество студентов и размеры комнаты.

**Выходные данные**<br />
Выведите три целых числа s, a1 и b1 (a ≤ a1; b ≤ b1) — итоговая площадь комнаты и ее размеры. Если существует несколько оптимальных решений, выведите любое.

### Выполнение:

```python
import math
#n, a, b = input().split(" ")
#num, side1, side2 = int(n), int(a), int(b)
def calc_area(num, side1, side2):
    #SQ = 1000000000000000000 
    SQ= 10e18
    if side1 * side2 < num * 6: #если введенной комнаты не хватает до минимальной установленной законом
        swap = False
        a1, b1 = 0, 0
        if side1 > side2: #сначала всегда должна идти меньшая, меняем местмаи и устанавливаем флаг
            side1, side2 = side2, side1
            swap = True
        for i in range (side1, int(math.ceil((math.sqrt(6*num))))): #проходимся по меньшей стороне от ее изначального значения до корня из оптимальной площади
            var = int(6 * num / i) #подбираем значения для второй стороны поделив оптимальную площадь на текущую меньшую сторону
            if var * i < 6 * num: var += 1 #если площадь с текущими сторонами меньше оптимальной, прибавляем +1 к незафиксированной стороне
            if var < side2: continue #если полученная сторона меньше исходной второй стороны, то пропускаем итерацию
            if var * i < SQ: #если площадь с текущими сторонами меньше 10^18 и прошлые условия не рсаботали, сохраняем значения
                SQ = var * i
                a1 = i
                b1 = var
        if swap: #если менялись местами стороны, то меняем местами сохраненные значения и выводим
            a1, b1 = b1, a1
        print(SQ)
        print(a1, b1)
    else: #если изначально введенная площадь уже больше или равна установленной законом, то просто выводим
        print (side1*side2)
        print(side1, side2)

calc_area(3, 3, 5) #Expected 18, 3, 6
calc_area(2, 4, 4) #Expected 16 4 4
calc_area(11, 7, 7) #Expected 70 7 10
calc_area(1, 1000000000, 1000000000) #Expected 1000000000000000000 1000000000 1000000000
```
