### Задание №2
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Написать генератор случайных матриц(многомерных), который принимает
опциональные параметры **m, n, min_limit, max_limit,** где **m** и **n** указывают размер
матрицы, а **min_lim** и **max_lim** - минимальное и максимальное значение для
генерируемого числа . По умолчанию при отсутствии параметров принимать следующие
значения: <br /> *m = 50 <br /> n = 50 <br /> min_limit = -250 <br /> max_limit = 1000 + 4 (номер своего варианта)*

```python
from random import seed
from random import randint

def createMatrix (): #Функция создания матрицы
    seed(1)
    m = input ('Enter columns ') #ввод данных пользователем
    n = input ('Enter rows ')
    min_limit = input ('Enter min value ')
    max_limit = input ('Enter max value ')
    matrix = []

    m = 50 if m == "" else  int (m) #Проверка введения пустой строки
    n = 50 if n == "" else  int (n)
    min_limit = -250 if min_limit == "" or int (max_limit) < int (min_limit) else int (min_limit) 
    max_limit = 1004 if max_limit == "" or int (max_limit) < int (min_limit) else int (max_limit)
    matrix = [ [0]* m for i in range(n) ] #выделение памяти под матрицу n раз - лист в листе шириной m

    for i in range(n): #заполнение случайными числами по сиду 1
        for j in range(m):
            matrix[i][j] = randint (min_limit, max_limit)
    return matrix


def printMatrix (matrix):       #Функция вывода матрицы
    for row in matrix:
        for elem in row:     
            print(elem, end = ' ')
        print()
matrix = createMatrix ()
printMatrix(matrix)
```

### Задание №3
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Реализовать методы сортировки строк числовой матрицы в соответствии с
заданием. Оценить время работы каждого алгоритма сортировки и сравнить его со
временем стандартной функции сортировки. Испытания проводить на сгенерированных
матрицах. <br /> Методы: <br /> 

| Выбором | Вставкой | Обменом | Шелла | Быстрая | Пирамидальная |
| ------- | -------- | ------- | ---- | ------- |-------------- |
| 0.012210 | 0.009177 | 0.024470 | 0.005876 | 0.001922 | 0.008822 | 

<br /> Время стандартного алгоритма: 0.000142 сек

```python
import datetime
import time
from random import seed
from random import randint

def createMatrix (): #Функция создания матрицы
    seed(1)
    m = input ('Enter columns ') #ввод данных пользователем
    n = input ('Enter rows ')
    min_limit = input ('Enter min value ')
    max_limit = input ('Enter max value ')
    matrix = []

    m = 50 if m == "" else  int (m) #Проверка введения пустой строки
    n = 50 if n == "" else  int (n)
    min_limit = -250 if min_limit == "" or int (max_limit) < int (min_limit) else int (min_limit)
    max_limit = 1004 if max_limit == "" or int (max_limit) < int (min_limit) else int (max_limit)
    matrix = [ [0]* m for i in range(n) ] #выделение памяти под матрицу n раз - лист в листе шириной m

    for i in range(n): #заполнение случайными числами по сиду 1
        for j in range(m):
            matrix[i][j] = randint (min_limit, max_limit)
    return matrix


def printMatrix (matrix):       #Функция вывода матрицы (для проверок)
    for row in matrix:
        for elem in row:     
            print(elem, end = ' ')
        print()

        
def selection (matrix): #Функция сортировки матрицы выбором. на вход матрица
    for row in range(len(matrix)): #цикл по количеству строк
        for i in range(len(matrix[0])): #цикл по количеству столбцов
            min = i #первый элемент строки назначается минимумумом

            for j in range (i + 1, len(matrix[0])): #поиск минимума в строке
                if matrix[row][j] < matrix[row][min]:
                    min = j
            matrix[row][i], matrix[row][min] = matrix[row][min], matrix[row][i]  #минимум сдвигается к началу


def insertion (matrix): #Функция сортировки матрицы вставкой. на вход матрица
    for row in range(len(matrix)):
        for i in range(1, len(matrix[0])):
            key = matrix[row][i]
            j = i-1
            while j >=0 and key < matrix[row][j] :
                matrix[row][j+1] = matrix[row][j]
                j -= 1
            matrix[row][j+1] = key 


def bubble (matrix):
    for row in range (len(matrix)):
        changed = True
        while changed:
            changed = False
            for i in range(len(matrix[0]) - 1):
                if matrix[row][i] > matrix[row][i+1]:
                    matrix[row][i], matrix[row][i+1] = matrix[row][i+1], matrix[row][i]
                    changed = True


def shell(matrix):
    for row in range (len(matrix)):
        inc = len(matrix[0]) 
        while inc:
            for i, el in enumerate(matrix[row]):
                while i >= inc and matrix[row][i - inc] > el:
                    matrix[row][i] = matrix[row][i - inc]
                    i -= inc
                matrix[row][i] = el
            inc = 1 if inc == 2 else int(inc * 5.0 / 11)


def quick(data):
	less = []
	pivotList = []
	more = []
	if len(data) <= 1:
		return data
	else:
		pivot = data[0]
		for i in data:
			if i < pivot:
				less.append(i)
			elif i > pivot:
				more.append(i)
			else:
				pivotList.append(i)
		less = quick(less)
		more = quick(more)
		return less + pivotList + more

    
def heapify(arr, n, i):
    largest = i  # Initialize largest as root
    l = 2 * i + 1     # left = 2*i + 1
    r = 2 * i + 2     # right = 2*i + 2
  
    # See if left child of root exists and is
    # greater than root
    if l < n and arr[i] < arr[l]:
        largest = l
  
    # See if right child of root exists and is
    # greater than root
    if r < n and arr[largest] < arr[r]:
        largest = r
  
    # Change root, if needed
    if largest != i:
        arr[i],arr[largest] = arr[largest],arr[i]  # swap
  
        # Heapify the root.
        heapify(arr, n, largest)

	
def heapSort(arr):
    n = len(arr)
  
    # Build a maxheap.
    # Since last parent will be at ((n//2)-1) we can start at that location.
    for i in range(n // 2 - 1, -1, -1):
        heapify(arr, n, i)
  
    # One by one extract elements
    for i in range(n-1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]   # swap
        heapify(arr, i, 0)

	  
matrix = createMatrix()
t = time.perf_counter()
selection(matrix)
print("Сортировка выбором {0:.6f} сек".format (time.perf_counter() - t))

matrix = createMatrix()
t = time.perf_counter()
insertion(matrix)
print("Сортировка вставкой {0:.6f} сек".format (time.perf_counter() - t))

matrix = createMatrix()
t = time.perf_counter()
bubble(matrix)
print("Сортировка обменом {0:.6f} сек".format (time.perf_counter() - t)) 

matrix = createMatrix()
t = time.perf_counter()
shell(matrix)
print("Сортировка шелл {0:.6f} сек".format (time.perf_counter() - t)) 

matrix = createMatrix()
t = time.perf_counter()
for row in range (len(matrix)):
    matrix[row] = quick(matrix[row])
print("Быстрая сортировка {0:.6f} сек".format (time.perf_counter() - t)) 

matrix = createMatrix()
t = time.perf_counter()
for row in range (len(matrix)):
    heapSort(matrix[row])
print("Пирамидальная сортировка {0:.6f} сек".format (time.perf_counter() - t)) 

matrix = createMatrix()
t = time.perf_counter()
for row in range (len(matrix)):
    matrix[row].sort
print("Стандартная сортировка {0:.6f} сек".format (time.perf_counter() - t)) 
```
