```sqlite3
import time
from random import seed
from random import randint

#Функция создания массива
def createArray ():
    seed(1)
    n = input ('Enter size ')
    min_limit = input ('Enter min value ')
    max_limit = input ('Enter max value ')
    n = 50 if n == "" else  int(n)
    min_limit = -250 if min_limit == "" or int (max_limit) < int (min_limit) else int (min_limit)
    max_limit = 1004 if max_limit == "" or int (max_limit) < int (min_limit) else int (max_limit)
    array = []
    for i in range(n):
            array.append(randint (min_limit, max_limit))
    return array

#Функция вывода массива
def printArray (array):       
    for elem in array:   
            print(elem, end = ' ')

#Функция бинарного поиска
def binarySearch(array, element):
    mid = 0
    start = 0
    end = len(array)
    step = 0

    while (start <= end):
        step = step+1 #увеличивается шаг
        mid = (start + end) // 2 #ищем медианный элемент

        if element == array[mid]: 
            return mid #если найден искомый

        if element < array[mid]:
            end = mid - 1 #если искомый элемент меньше медианного
        else:
            start = mid + 1 #иначе
    return -1 #если не найден

print("Бинарный поиск")
array = createArray()
array.sort()
printArray(array)
element = input('Choose element: ')
t = time.perf_counter()
print("Индекс:", binarySearch(array, int(element)))
print("Бинарный поиск: {0:.6f} сек".format (time.perf_counter() - t))
t = time.perf_counter()
find = array.index(int(element))
print("Встроенный поиск: {0:.6f} сек".format ((time.perf_counter() - t)))
print (find)
```

```sqlite3
#Класс узла дерева
class Node:
    def __init__(self, key):
        self.left = None #Ссылка на левую ветку
        self.right = None #Ссылка на правую ветку
        self.val = key #Значение корня узла
 
#Функция заполнения дерева
def insert(root, key):
    if root is None: #Если корень пуст
        return Node(key) #возвращает узел дерева с указанным корнем
    else:
        if root.val == key: #если такой элемент уже есть
            return root
        elif root.val < key: #если ключ больше корня, то значение вставляется в правую ветку ( там новый корень)
            root.right = insert(root.right, key)
        else:
            root.left = insert(root.left, key)  #если ключ меньше корня, то значение вставляется в левую ветку ( там новый корень)
    return root

#поиск по бинарному дереву
def searchBT(root,key): 
    if root is None or root.val == key: #если значение нашлось в корне или корень пуст
        return root
    if root.val < key: #если искомое значение больше коренного, то дальше поиск идет в правой ветке
        return searchBT(root.right,key)
    return searchBT(root.left,key) #если меньше, то поиск идет в левой ветке

#заполнение бинарного дерева указанным количеством случайных чисел
def createBT (size): 
    seed(time.perf_counter)
    r = Node(randint (-250, 1004))
    for i in range(size - 1):
            r = insert(r, randint (-250, 1004))
    return r

#функция вывода дерева
def inorder(root):
    if root:
        inorder(root.left)
        print(root.val)
        inorder(root.right)

print("поиск по бинарному дереву")
r = createBT(50)
inorder(r)
key = input("what key should be found ")
t = time.perf_counter()
res = searchBT(r, int(key))
print("Поиск по бинарному дереву: {0:.6f} сек".format (time.perf_counter() - t))
if res is not None:
    print(res, res.val)
```

```sqlite3
#Интерполяционный поиск 
def interpolationSearch(arr, val):
    low = 0 #нижняя граница
    high = (len(arr) - 1) #верхняя граница
    while low <= high and val >= arr[low] and val <= arr[high]: #пока искомое значение находится в указанных рамках
        index = low + int(((float(high - low) / ( arr[high] - arr[low])) * ( val - arr[low]))) #формула вероятностная дял нахождения возможного индекса искомого элемента
        if arr[index] == val:
            return index #если найден
        if arr[index] < val:
            low = index + 1 #если значение по вероятному индексу меньше искомого, то нижняя граница сдвигается вправо
        else:
            high = index - 1 #если значение по вероятному индексу больше искомого, то верхняя граница сдвигается влево
    return -1 #если элемент не найден

print("Интерполяционный поиск")
array = createArray()
array.sort()
printArray(array)
key = input("what key should be found ")
t = time.perf_counter()
res = interpolationSearch(array, int(key))
print("Интерполяционный поиск: {0:.6f} сек".format (time.perf_counter() - t))
print(res)
```

```sqlite3
class Node:
    #Лист значений для каждого ключа
    def __init__(self, value):
        self.value = value
        self.next = None
# словарь ключ - значение
class HashDict:
    # Конструктор инициализации
    def __init__(self, d):
        self.d = d
        self.size = len(self.d)

    # Хэш-функция для ключа в виде строки
    def hash(self, key):
        hash_key = 0
        for i in range (len(key)):
            hash_key= hash_key * 10 + ord(key[i])
        return hash_key
        
#Метод вставки пары ключ-значение в словарь
    def insert(self, key, value):
        self.size += 1
        index = self.hash(key)
        node = self.d.get(index)
        if node is None: #Если такого узла нет, то добавляем ключ-значение
            self.d[index] = Node(value) #значение записывается как объект формата списка
        else:
            # Когда начальная точка связанного списка уже занята, необходимо создать новый узел для хранения данных
            temp = Node(value)
            #p указывает на головной узел связанного списка, начиная с [позиции, где нижний индекс является значением хеш-функции в хеш-таблице].
            p = self.d.get(index)
            # Проверять поочередно назад
            while p.next != None:
                p = p.next
            p.next = temp
    #Метод вывода содержимого словаря
    def print (self):
        x = list(self.d.values())
        for i in range(len(x)):
            temp = x[i]
            while temp is not None: #Пока значения списка не кончатся
                print(temp.value, end=' ') 
                temp = temp.next #Следующий элемент списка
            print(' ')

    #Метод поиска элемента по ключу
    def get(self, key):
        index = self.hash(key) # получение хэш-кода для ключа
        node = self.d.get(index)
        if node is not None:  
            res = [] #список для вывода значений, если по одному ключу находятся несколько значений
            res.append(node.value)
            while node.next is not None:
                node = node.next
                res.append(node.value)
            return res	
        else:
            return None


print("рехэширование списками")
dict = {}
ht = HashDict(dict)
ht.insert("petya", 101)
ht.insert("vasya", 22)
ht.insert("vasya", 122)
ht.insert("jenya", 33)
ht.insert("helga", 44)
ht.insert("helga", 55)
ht.insert("helga", 66)
ht.print()
print(', '.join(map(str,(ht.get("helga") ) ) ) )
```
